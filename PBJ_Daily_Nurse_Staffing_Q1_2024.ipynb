```python
import pandas as pd

df = pd.read_csv('PBJ_Nursingstaffing.csv', encoding='ISO-8859-1')

df.head()
```

```python
df.isnull().sum()

missing_percentage = df.isnull().sum() / len(df) * 100
print(missing_percentage)
```

```python
print(df.columns)
```

```python
df_cleaned = df.copy()
```

```python
# Define the list of relevant columns (employee and contractor hours)
hours_columns = [
    'Hrs_RN_emp', 'Hrs_RN_ctr',
    'Hrs_LPN_emp', 'Hrs_LPN_ctr',
    'Hrs_CNA_emp', 'Hrs_CNA_ctr',
    'Hrs_MedAide_emp', 'Hrs_MedAide_ctr'
]

# Fill missing values in those columns with 0
df_cleaned[hours_columns] = df_cleaned[hours_columns].fillna(0)
```

```python
# Calculate total employee hours by summing across different roles
df_cleaned['Total_Employee_Hours'] = (
    df_cleaned['Hrs_RN_emp'] + df_cleaned['Hrs_LPN_emp'] +
    df_cleaned['Hrs_CNA_emp'] + df_cleaned['Hrs_MedAide_emp']
)

# Calculate total contractor hours by summing across different roles
df_cleaned['Total_Contractor_Hours'] = (
    df_cleaned['Hrs_RN_ctr'] + df_cleaned['Hrs_LPN_ctr'] +
    df_cleaned['Hrs_CNA_ctr'] + df_cleaned['Hrs_MedAide_ctr']
)
```

```python
# Calculate contractor-to-employee ratio
df_cleaned['contractor_ratio'] = df_cleaned['Total_Contractor_Hours'] / df_cleaned['Total_Employee_Hours']

# Preview the result
df_cleaned[['Total_Employee_Hours', 'Total_Contractor_Hours', 'contractor_ratio']].head()
```

```python
# Example: Group by state and calculate the average contractor ratio per state
state_contractor_usage = df_cleaned.groupby('STATE')['contractor_ratio'].mean().reset_index()

# Visualize contractor usage by state
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(state_contractor_usage['STATE'], state_contractor_usage['contractor_ratio'])
plt.title('Average Contractor Ratio by State')
plt.xlabel('State')
plt.ylabel('Contractor Ratio')
plt.xticks(rotation=90)
plt.show()
```

A. State-Level Analysis
Objective: Identify states with higher reliance on contractor staffing. This can help prioritize sales efforts in states where contractor demand is high.

```python
# Group by state and calculate the average contractor ratio per state
state_contractor_usage = df_cleaned.groupby('STATE')['contractor_ratio'].mean().reset_index()

# Sort by contractor ratio to identify the states with the highest reliance on contractors
state_contractor_usage_sorted = state_contractor_usage.sort_values(by='contractor_ratio', ascending=False)

# Preview the top states
state_contractor_usage_sorted.head()
```

```python
# Replace 'inf' values with NaN and then drop them
state_contractor_usage_sorted.replace([float('inf'), -float('inf')], pd.NA, inplace=True)

# Drop rows with NaN (previously inf)
state_contractor_usage_sorted = state_contractor_usage_sorted.dropna()

# Sort by contractor ratio again to ensure correct ordering
state_contractor_usage_sorted = state_contractor_usage_sorted.sort_values(by='contractor_ratio', ascending=False)

# Preview the cleaned data
state_contractor_usage_sorted.head()
```

```python
# Bar plot for contractor ratio by state
plt.figure(figsize=(10, 6))
plt.bar(state_contractor_usage_sorted['STATE'], state_contractor_usage_sorted['contractor_ratio'])
plt.title('Average Contractor-to-Employee Ratio by State')
plt.xlabel('State')
plt.ylabel('Contractor Ratio')
plt.xticks(rotation=90)
plt.show()
```

Recommendation: You can recommend that the sales team focuses on the top states with high contractor ratios. These states have more facilities relying on contractors, which could be good prospects for Clipboard Health’s services.

Facility-Level Analysis
Objective: Identify specific facilities with a high reliance on contractors.

```python
# Identify top 10 facilities by contractor-to-employee ratio
top_facilities = df_cleaned[['PROVNUM', 'PROVNAME', 'STATE', 'Total_Contractor_Hours', 'Total_Employee_Hours', 'contractor_ratio']]

# Replace 'inf' and '-inf' values with NaN and drop them
top_facilities.replace([float('inf'), -float('inf')], pd.NA, inplace=True)
top_facilities = top_facilities.dropna()

# Sort by contractor ratio and get the top 10 facilities
top_facilities = top_facilities.sort_values(by='contractor_ratio', ascending=False).head(10)

# Preview the top facilities
top_facilities
```

```python
# Bar plot for top facilities by contractor ratio
top_facilities_plot = top_facilities[['PROVNAME', 'contractor_ratio']].set_index('PROVNAME')

plt.figure(figsize=(10, 6))
top_facilities_plot.plot(kind='bar')
plt.title('Top 10 Facilities by Contractor-to-Employee Ratio')
plt.ylabel('Contractor Ratio')
plt.xlabel('Facility Name')
plt.xticks(rotation=90)
plt.show()
```

Recommendation: These facilities represent high-potential clients who may benefit from more contractor staffing solutions. You can recommend these facilities as potential leads.

Contractor Usage by Role (RN, LPN, CNA)
Objective: Break down contractor usage by specific roles (Registered Nurses, Licensed Practical Nurses, Certified Nursing Assistants, etc.).


```python
# Calculate average contractor hours by role (RN, LPN, CNA) for each state
contractor_role_usage = df_cleaned.groupby('STATE')[
    ['Hrs_RN_ctr', 'Hrs_LPN_ctr', 'Hrs_CNA_ctr', 'Hrs_MedAide_ctr']
].mean().reset_index()

# Clearer breakdown for each role:
# 1. Registered Nurses (RN)
top_rn_states = contractor_role_usage[['STATE', 'Hrs_RN_ctr']].sort_values(by='Hrs_RN_ctr', ascending=False)
print("Top States for RN Contractor Usage:")
print(top_rn_states.head(5))  # Display the top 5 states for RN contractors

# 2. Licensed Practical Nurses (LPN)
top_lpn_states = contractor_role_usage[['STATE', 'Hrs_LPN_ctr']].sort_values(by='Hrs_LPN_ctr', ascending=False)
print("\nTop States for LPN Contractor Usage:")
print(top_lpn_states.head(5))  # Display the top 5 states for LPN contractors

# 3. Certified Nursing Assistants (CNA)
top_cna_states = contractor_role_usage[['STATE', 'Hrs_CNA_ctr']].sort_values(by='Hrs_CNA_ctr', ascending=False)
print("\nTop States for CNA Contractor Usage:")
print(top_cna_states.head(5))  # Display the top 5 states for CNA contractors

# 4. Med Aides (if needed)
top_medaide_states = contractor_role_usage[['STATE', 'Hrs_MedAide_ctr']].sort_values(by='Hrs_MedAide_ctr', ascending=False)
print("\nTop States for Med Aide Contractor Usage:")
print(top_medaide_states.head(5))  # Display the top 5 states for Med Aide contractors
```

```python
# Bar plot for contractor usage by role
contractor_role_usage_plot = contractor_role_usage.set_index('STATE')

contractor_role_usage_plot.plot(kind='bar', figsize=(12, 8))
plt.title('Average Contractor Usage by Role (RN, LPN, CNA) for Each State')
plt.xlabel('State')
plt.ylabel('Average Contractor Hours')
plt.xticks(rotation=90)
plt.show()
```

Recommendation: If you see higher contractor usage for specific roles (e.g., RNs or CNAs), you can recommend targeting facilities with staffing gaps in these roles.
